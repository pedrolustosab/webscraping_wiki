```python
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Pagina e tema 
url = "https://www.wikipedia.org/"
tema = "copa do mundo 2014"
```

```python
# Inicializa o driver
driver = webdriver.Chrome()

# Abre a página do Wikipedia
driver.get(url)

# Espera a caixa de busca carregar e insere o termo de busca
search_box = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.NAME, "search"))
)
search_box.send_keys(tema)
search_box.send_keys(Keys.RETURN)  

# Tratar tipo de pesquisa
if driver.find_elements(By.XPATH,'//*[@id="disambig"]/table/tbody/tr/td[1]/span/a/img'):
    driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/ul/li[1]/a')[0].click()
elif driver.find_elements(By.XPATH,'//*[@id="firstHeading"]'):
    if driver.find_elements(By.XPATH,'//*[@id="firstHeading"]')[0].text == "Resultados da pesquisa":
        driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[3]/div[4]/ul/li[1]/div[2]/div[2]/div[1]/a')[0].click()
    else:
        pass

# Obtém o HTML da página
html = driver.page_source

# Fecha o driver
driver.quit()
```

```python
# Imprime Tabela
html_parsed = BeautifulSoup(html, "html.parser")
tabelas = html_parsed.find_all("table")
for i, tabela in enumerate(tabelas):
    df = pd.read_html(str(tabela))[0]
    print(df.head())
    df.to_csv(f"tabela_{i + 1}.csv", index=False)
```
